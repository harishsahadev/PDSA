## Compute primes from 1 to 100

```python
def factors(n):
  factorlist = []
  for i in range(1, n+1):
    if n%i == 0:
      factorlist.append(i)
  return factorlist
```

```python
def prime(n):
  return (factors(n) == [1,n])
```

```python
def prime(n):
  return len(factors(n)) == 2
```

```python
primelist = []
for i in range(1, 101):
  if prime(i):
    primelist.append(i)

print(primelist)
```

# L 1.4

## Counting Primes

```python
# List of all primes upto m

def primesupto(m):
  pl = [] # prime list
  for i in range(1, m+1):
    if prime(i):
      pl.append(i)
  return pl
```

```python
primesupto(70)
```

```python
# First m primes

def firstprimes(m):
  (count, i, pl) = (0, 1, [])
  while (count < m):
    if prime(i):
      (count, pl) = (count+1, pl+[i])
    i = i + 1
  return pl
```

```python
firstprimes(10)
```

```python
# Better way to find prime

import math
def prime(n):
  (result, i) = (True, 2)
  while (result and (i <= math.sqrt(n))):
    if (n%i) == 0:
      result = False
    i = i+1
  return result
```

## Computing GCD in a different way

Suppose d divides m and n
- m = ad, n = bd
- m-n = (a-b)d
- Thus, d also divides m-n

This takes time proportional to max(m,n)

```python
def gcd(m, n):
  (a,b) = (max(m,n), min(m,n))
  if a%b == 0:
    return b
  else:
    return gcd(b, a-b)
```

```python
gcd(144, 132)
```

## GCD using Euclid's Algorithm

Suppose n doesn't divide m
- Then **m = qn + r** ; q : quotient and r : remainder
- Suppose d divides both m and n
- m = ad, n = bd
- Thus, ad = q(bd) + r
- r = d (a-qb)
- So r must be of the form cd

```python
def gcd(m, n):
  (a,b) = (max(m,n), min(m,n))
  if a%b == 0:
    return b
  else:
    return gcd(b, a%b)
```

## Class example : 2D Points

Point:
- coordinates (x,y)
- Traslate : Shift a point by (x, y)
- odistance : Distance from origin
  - $d = \sqrt{x^2 + y^2}$
- \_\_str__() : convert object into string
  - Implicitly invoked by **print()**
- \_\_add__() : Invoked by **+**

```python
class Point:
  def __init__(self, a=0, b=0):
    self.x = a
    self.y = b

  def translate(self, deltax, deltay):
    self.x += deltax
    self.y += deltay

  def odistance(self):
    import math
    d = math.sqrt(self.x*self.x + self.y*self.y)
    return d

  def __str__(self):
    return ('(' + str(self.x) + ',' + str(self.y) + ')')

  def __add__(self, p):
    return (Point(self.x + p.x,
                  self.y + p.y))
```

## Timing Code
- To find the speed of Python using Timer Class
- Using inbuilt library **time** using the function **perf_time()** which is a performance counter
  - Compare 2 consecutive readings to get an interval 
  - Default in secs
- Python executes **$10^7$** operations per second

```python
# Timer Class
import time

class TimerError(Exception):
  """A custom exception used to report errors in Timer Class""" # This is the documentation of the class/error

class Timer:
  def __init__(self):
    self.startTime = None
    self.elapsedTime = None

  def start(self):
    """Start a new timer""" # Can see this documentation when hovering over the function
    if self.startTime is not None:
      raise TimerError("Timer is running. Use .stop()")
    self.startTime = time.perf_counter()

  def stop(self):
    """Save the elapsed time and re-initialize timer"""
    if self.startTime is None:
      raise TimerError("Timer is not running. Use .start()")
    self.elapsedTime = time.perf_counter() - self.startTime
    self.startTime = None

  def elapsed(self):
    """Report elapsed time"""
    if self.elapsedTime is None:
      raise TimerError("Timer has not been run yet. Use .start()")
    return self.elapsedTime

  def __str__(self):
    """print() prints elapsed time"""
    return str(self.elapsedTime)
```

###Code to check python's performance (time) in performing $10^4$, $10^5$, ... $10^8$ operations using the Timer class

```python
t = Timer()
for j in range(4, 9):
  t.start()
  n = 0
  for i in range(10**j):
    n = n + i
  t.stop()
  print(j, t)
```
